```python
#import turtle library as t to draw the art
import turtle as t

#Import colorsys as c to convert colors from hsv to rgb
import colorsys as c

#Import random library to generate random integers
import random

#Set the size of the window width and height=800
t.setup(800,800)

#Set speed of the pen range should be in 0 to 10
#Fastest value is '0', slowest value is '1'
t.speed(0)

#Set the thickness of the pen to 5
t.width(5)

#set the background of the window to black
t.bgcolor("black")


# Define a function to draw a star
def star():
    
     #Set the color of star to yellow
    t.color("yellow")
    
    #Starting to fill the shape with color yellow
    t.begin_fill()  
    
    for i in range(5): 
        
        #Drawing the shape by going forward by 10px
        t.forward(10)  
        
        #Drawing the shape by going to right side by 144px
        t.right(144)
        
    #Stoping the filling the shape with color yellow
    t.end_fill()


# Draw the first star on bottom left corner
#Moving the turtle to new place without drawing a line on the screen
t.penup()

# bottom left
t.goto(-350, -350)  

#Allowing the turtle to draw on the surface
t.pendown()

#calling for the function
# star()



# Draw the second star on top right corner
#Moving the turtle to new place without drawing a line on the screen
t.penup()

# top right
t.goto(350, 350) 

#Allowing the turtle to draw on the surface
t.pendown()

#calling for the function
# star()



# Draw the third star on bottom right corner
#Moving the turtle to new place without drawing a line on the screen
t.penup()

# bottom right
t.goto(350, -350)  

#Allowing the turtle to draw on the surface
t.pendown()

#calling for the function
# star()



# Draw the fourth star on top left corner
#Moving the turtle to new place without drawing a line on the screen
t.penup()

# top left
t.goto(-350, 350)  

#Allowing the turtle to draw on the surface
t.pendown()

#calling for the function
# star()



# Draw the bottom middle star
#Moving the turtle to new place without drawing a line on the screen
t.penup()

 # bottom 
t.goto(0, -350) 

#Allowing the turtle to draw on the surface
t.pendown()

#calling for the function
# star()



# Draw the top middle star
#Moving the turtle to new place without drawing a line on the screen
t.penup()

# top 
t.goto(0, 350) 

#Allowing the turtle to draw on the surface
t.pendown()

#calling for the function
# star()



# Draw the right middle star
#Moving the turtle to new place without drawing a line on the screen
t.penup()

# right
t.goto(-350, -0) 

#Allowing the turtle to draw on the surface
t.pendown()

#calling for the function
# star()



# Draw the left middle star
#Moving the turtle to new place without drawing a line on the screen
t.penup()

# left 
t.goto(350, 0)  

#Allowing the turtle to draw on the surface
t.pendown()

#calling for the function
# star()



#Moving the turtle to new place without drawing a line on the screen
t.penup()

#Set the position to center to draw the flower
t.goto(0,-40)

#Allowing the turtle to draw on the surface
t.pendown()

#Randomly pick up the integers between 5 to 30
r1=random.randint(5,30)

#Randomly pick up the integers between 5 to 20
r2=random.randint(5,20)

#Randomly pick up the integers between 15 to 30
r3=random.randint(15,30)

#Set the color gradients of the petals randomly
for j in range(r1):
    
    #Decides the height of the petals randomly
    for i in range(r2):
        
        #convert hsv colors to rgb colors get various color shades
        t.color(c.hsv_to_rgb(i/15,j/25,1))
        
        #Turns the turtle clockwise
        t.right(90)
        
        #Draw a circle with radius 200-j*4 and extent of the circle to 90 
        t.circle(200-j*4,90)
        
        #Turns the turtle counterclockwise
        t.left(90)
        
        t.circle(200-j*4,90)
        t.right(180)
        
        #Draw the main circle with 50 radius and divide it into random parts, random petals would be drawn
        t.circle(50,r3)
        
        
# Hide the turtle when finished
t.hideturtle()

#Keep the window open until user closes it
t.done()
```

```python
import turtle
import matplotlib.pyplot as plt
import numpy as np

# Generate 7 colors from the rainbow colormap
colors = plt.cm.rainbow(np.linspace(0, 1, 7))

# Function to convert RGBA to Hex
def rgba_to_hex(rgba):
    r, g, b, _ = rgba
    return "#{:02x}{:02x}{:02x}".format(int(r*255), int(g*255), int(b*255))

# Convert colors to hex format
hex_colors = [rgba_to_hex(color) for color in colors]

# Initialize the turtle
t = turtle.Pen()
turtle.bgcolor('black')

# Draw with the turtle using the rainbow colors
for x in range(200):
    t.pencolor(hex_colors[x % 7])
    t.width(x / 100 + 1)
    t.forward(x)
    t.left(59)

# Finish the drawing
turtle.done()
```
